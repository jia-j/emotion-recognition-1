## Visualizing the MAHNOB_HCI dataset

```python
import pyedflib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sample = "/Users/Russel/myProjects/emotion-recognition/code/datasets/mahnob_hci/Sessions/522/Part_5_S_Trial1_emotion.bdf"
```

```python
f = pyedflib.EdfReader(sample)
n = f.signals_in_file
n
```

```python
signal_labels = f.getSignalLabels()
signal_labels, len(signal_labels)
```

```python
sigbufs = np.zeros((n, f.getNSamples()[0]))
for i in np.arange(n):
    print(i, signal_labels[i])
    sigbufs[i, :] = f.readSignal(i)
```

```python
sigbufs.shape
```

```python
ECG = [[], [], []]
ECG[0] = sigbufs[32].reshape(-1, 1)
ECG[1] = sigbufs[33].reshape(-1, 1)
ECG[2] = sigbufs[34].reshape(-1, 1)
```

```python
norm_ECG = ECG[2][:2000]
from matplotlib.pyplot import figure
figure(num=None, figsize=(20, 8), dpi=80, facecolor='w', edgecolor='k')
plt.plot(norm_ECG)
```
